```python
import pandas as pd
import numpy as np
```

```python
from functools import reduce
```

```python
df = pd.read_excel('/content/Population Density.xlsx', index_col = 0, skiprows=[0,1,3])
df
```

```python
def melt_df(df):
    df_ = df.copy()
    index = df_.index
    df_ = df_.transpose().reset_index().melt(id_vars= 'index', value_vars = index)
    
    return df_
```

```python
df = melt_df(df)
```

```python
df['index'] = df['index'].str.rstrip(' (*)')
df.columns = ['Year', 'Province', 'Population Density']
```

```python
df
```

```python
df1 = pd.read_excel('average.xlsx', skiprows = [0,1,2], index_col = 0)
df1.index = df1.index.str.strip()
df1
```

```python
attributes = np.array_split(df1.columns, 5)
```

```python
df1_1 = melt_df(df1[attributes[0]])
df1_1['index'] = df1_1['index'].str.lstrip('Prel. ')
index = df1_1['index']
df1_1.columns = ['Year', 'Province', 'Average Pop']
df1_1
```

```python
df1_2 = melt_df(df1[attributes[1]])
df1_2['index'] = index
df1_2.columns = ['Year', 'Province', 'Average Pop (Male)']
df1_2
```

```python
df1_3 = melt_df(df1[attributes[2]])
df1_3['index'] = index
df1_3.columns = ['Year', 'Province', 'Average Pop (Female)']
df1_3
```

```python
df1_4 = melt_df(df1[attributes[3]])
df1_4['index'] = index
df1_4.columns = ['Year', 'Province', 'Average Pop (Urban)']
df1_4
```

```python
df1_5 = melt_df(df1[attributes[4]])
df1_5['index'] = index
df1_5.columns = ['Year', 'Province', 'Average Pop (Rural)']
df1_5
```

```python
df2 = pd.read_excel('Sex Ratio.xlsx', skiprows = [0,1], index_col = 0)
df2 = melt_df(df2)
df2.columns = ['Year', 'Province', 'Sex Ratio']
df2.Year = df2.Year.str.lstrip('Prel. ')
df2.Province = df2.Province.str.strip()
df2
```

```python
df3 = pd.read_excel('Ageing_Index.xlsx', skiprows = [0,1,2,3,4,5], index_col = 0, usecols = "B:G")
df3 = melt_df(df3)
df3.fillna('WHOLE COUNTRY', inplace = True)
df3['index'] = df3['index'].str.lstrip('YEAR ')
df3.columns = ['Year', 'Province', 'Ageging Index']
df3
```

```python
df4 = pd.read_excel('Fertility Rate.xlsx', skiprows = [0,1], index_col=0)
df4 = melt_df(df4)
df4.columns = ['Year', 'Province', 'Fertility Rate']
df4.Year = df4.Year.str.lstrip('Prel. ')
df4.Province = df4.Province.str.strip()
df4
```

```python
df5 = pd.read_excel('Crude Birth Rate.xlsx', skiprows = [0,1,2], index_col=0)
df5 = melt_df(df5)
df5['index'] = df5['index'].str.strip()
df5.columns = ['Year', 'Province', 'Crude Birth Rate']
df5.Province = df5.Province.str.strip()
df5.Year = df5.Year.str.lstrip('Prel. ')
df5
```

```python
df6 = pd.read_excel('Life Expectancy.xlsx', skiprows = [0,1], index_col=0)
df6 = melt_df(df6)
df6.columns = ['Year', 'Province', 'Life Expectency']
df6.Year = df6.Year.str.lstrip('Prel. ')
df6.Province = df6.Province.str.strip()
df6.replace('North Central and Central coastal areas', 'Northern Central area and Central coastal area', inplace = True)
df6
```

```python
dfs =[df1_1, df1_2, df1_3, df1_4, df1_5, df, df2, df3, df4, df5, df6]
df_final = reduce(lambda left,right: pd.merge(left,right,on=['Year', 'Province'], how = 'outer'), dfs)
df_final
```

```python
df_final.to_csv('dataset.csv', index = True)
```

```python
df7 = pd.read_excel('Productivity.xlsx', skiprows = [0,1], index_col=0)
df7 = melt_df(df7)
df7.columns = ['Year', 'Industry', 'Productivity']
df7.Year = df7.Year.str.lstrip('Prel. ')
df7
```

```python
df8 = pd.read_excel('Trained Workers.xlsx', skiprows = [0,1], index_col=0)
df8 = melt_df(df8)
df8.columns = ['Year', 'Industry', 'Trained Workers']
df8.Year = df8.Year.str.lstrip('Prel. ')
df8
```

```python
df9 = pd.read_excel('/content/Undereployment Rate.xlsx', skiprows = [0,1], index_col=0, usecols = "A, C:M")
df9 = melt_df(df9)
df9.columns = ['Year', 'Industry', 'Undereployment Rate']
df9.replace('WHOLE COUNTRY', 'TOTAL', inplace = True)
df9.Year = df9.Year.str.lstrip('Prel. ')
df9
```

```python
df10 = pd.read_excel('/content/Unemployment Rate.xlsx', skiprows = [0,1], index_col=0, usecols = "A, C:M")
df10 = melt_df(df10)
df10.columns = ['Year', 'Industry', 'Unemployment Rate']
df10.replace('WHOLE COUNTRY', 'TOTAL', inplace = True)
df10.Year = df10.Year.str.lstrip('Prel. ')
df10
```

```python
df11 = pd.read_excel('/content/Qualification.xlsx', skiprows = [0, 1], index_col=0)
df11 = melt_df(df11)
df11.columns = ['Year', 'Industry', 'Qualification Rate']
df11.Year = df11.Year.str.lstrip('Prel. ')
df11
```

```python
dfs =[df7, df8, df9, df10, df11]
df_final1 = reduce(lambda left,right: pd.merge(left,right,on=['Year', 'Industry'], how = 'outer'), dfs)
df_final1
```

```python
df_final1.to_csv('Economics.csv', index = False)
```
